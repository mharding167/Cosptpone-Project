### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
df.head()
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here

# Display data types and non-null counts for each column
print("Data Types and Non-Null Counts:")
df.info()

# Display missing values for each column
print("\nMissing Values per Column:")
missing_values = df.isnull().sum()
print(missing_values)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here

# Generate basic statistics for numerical columns
numerical_stats = df.describe()

# Display the statistics
print(numerical_stats)
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here


# Check for empty or null values
empty_or_null = df[df['Country'].isnull() | (df['Country'] == '')]

#Check for leading or trailing spaces
leading_trailing_spaces = df[df['Country'].str.strip() != df['Country']]

# Check for irrelevant entries (common placeholders)
irrelevant_entries = df[df['Country'].isin(['N/A', 'None', 'Unknown', 'Unspecified', 'Other'])]

# 5. Check for inconsistent capitalization
inconsistent_capitalization = df[df['Country'] != df['Country'].str.title()]

# Print the inconsistent or irrelevant entries found in each category
print("Empty or null entries:\n", empty_or_null['Country'], "\n")
print("Entries with leading/trailing spaces:\n", leading_trailing_spaces['Country'], "\n")
print("Entries with irrelevant placeholders:\n", irrelevant_entries['Country'], "\n")
print("Entries with inconsistent capitalization:\n", inconsistent_capitalization['Country'], "\n")
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here

# Standardize Country column by mapping inconsistent entries to a consistent format
country_mapping = {
    ' USA': 'USA', 
    'United States': 'USA',
    'United States of America': 'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'UK',
    'Iran, Islamic Republic of Iran': 'Iran'
}

df['Country'] = df['Country'].replace(country_mapping)

print(df['Country'])
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here

df_encoded = pd.get_dummies(df, columns=['Employment'], drop_first=False)
df_encoded.head()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here

# Count missing values for each column
missing_values = df.isnull().sum()

# Sort the columns by the number of missing values (in descending order)
missing_values_sorted = missing_values.sort_values(ascending=False)

# Print the columns with the highest number of missing values
print(missing_values_sorted)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here

# Identify numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Impute missing values with the mean for all numerical columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Print the updated df
print(df)
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here

# Identify categorical columns with object data type
categorical_columns = df.select_dtypes(include=['object']).columns

# Impute missing values in categorical columns with the most frequent value
for col in categorical_columns:
    most_frequent_value = df[col].mode()[0]  
    df[col] = df[col].fillna(most_frequent_value)

# 3. Print the updated DataFrame
print(df)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here

# min and max values of the 'ConvertedCompYearly' column
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()

# Apply Min-Max scaling
df['ConvertedCompYearly_scaled'] = (df['ConvertedCompYearly'] - min_value) / (max_value - min_value)

# Print the updated df with the normalized column
print(df)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here

import numpy as np

# Apply log transformation to the 'ConvertedCompYearly' column
df['ConvertedCompYearly_log'] = np.log(df['ConvertedCompYearly'])

# Print the updated DataFrame with the log-transformed column
print(df)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here

# Function to assign Experience Level based on YearsCodePro
def assign_experience_level(years):
    if years <= 2:
        return 'Beginner'
    elif 3 <= years <= 5:
        return 'Intermediate'
    else:
        return 'Advanced'

# Apply the function to create the 'ExperienceLevel' column
df['ExperienceLevel'] = df['YearsCodePro'].apply(assign_experience_level)

# Print the updated DataFrame
print(df['ExperienceLevel'])
```
