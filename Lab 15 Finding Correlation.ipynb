##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here


# Set up the plot style
plt.figure(figsize=(12, 6))

# Create a histogram with KDE
sns.histplot(data=df, x='ConvertedCompYearly', kde=True, color='skyblue')

# Customize the plot
plt.title('Distribution of Yearly Compensation', fontsize=16)
plt.xlabel('Yearly Compensation ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add median line
median = df['ConvertedCompYearly'].median()
plt.axvline(median, color='red', linestyle='dashed', linewidth=2, label=f'Median: ${median:,.0f}')

# Add mean line
mean = df['ConvertedCompYearly'].mean()
plt.axvline(mean, color='green', linestyle='dashed', linewidth=2, label=f'Mean: ${mean:,.0f}')

plt.legend()

# Limit x-axis to focus on the main distribution (e.g., up to 99th percentile)
upper_limit = df['ConvertedCompYearly'].quantile(0.99)
plt.xlim(0, upper_limit)

# Show the plot
plt.tight_layout()
plt.show()

# Print some summary statistics
print(df['ConvertedCompYearly'].describe())
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here

# Filter the DataFrame for "Employed, full-time"
full_time_employees = df[df['Employment'] == 'Employed, full-time']

# Calculate the median compensation for the filtered data
median_compensation = full_time_employees['ConvertedCompYearly'].median()

# Print the result
print(f"Median compensation for 'Employed, full-time' respondents: {median_compensation}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here

# Replace missing values (NaN) in the 'ConvertedCompYearly' column with the mean

# Create a box plot to compare compensation distributions by country
plt.figure(figsize=(18, 10))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df)

# Add title and labels
plt.title('Compensation Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Converted Annual Compensation')
plt.xticks(rotation=90, ha='right')

# Show the plot
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here

# Calculate Q1, Q3, and IQR for ConvertedCompYearly
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = max(0, Q1 - 1.5 * IQR)
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['ConvertedCompYearly'] < lower_bound) | (df['ConvertedCompYearly'] > upper_bound)]

# Print summary statistics
print(f"Q1: ${Q1:.2f}")
print(f"Q3: ${Q3:.2f}")
print(f"IQR: ${IQR:.2f}")
print(f"Lower bound: ${lower_bound:.2f}")
print(f"Upper bound: ${upper_bound:.2f}")
print(f"\nNumber of outliers: {len(outliers)}")
print(f"Percentage of outliers: {(len(outliers) / len(df)) * 100:.2f}%")

# Plotting the box plot to visualize the outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['ConvertedCompYearly'], color='lightblue')
plt.title('Box Plot for Yearly Compensation with Outliers')
plt.xlabel('Yearly Compensation')
plt.show()

# Create new DataFrame excluding outliers

# Replace missing values (NaN) in the 'ConvertedCompYearly' column with the mean
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(mean)

df_no_outliers = df[(df['ConvertedCompYearly'] > lower_bound) & (df['ConvertedCompYearly'] < upper_bound)]
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>



**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here


# Calculate the correlation matrix between 'ConvertedCompYearly', 'WorkExp', and 'JobSatPoints_1'
correlation_matrix = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)

# Add title and labels
plt.title('Correlation Heatmap between Compensation, Work Experience, and Job Satisfaction')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here

# Set up the plot size
plt.figure(figsize=(15, 8))

# Create scatter plot for 'ConvertedCompYearly' vs. 'WorkExp'
plt.subplot(1, 2, 1)
sns.scatterplot(x='WorkExp', y='ConvertedCompYearly', data=df, color='green')
plt.title('Converted Compensation vs. Work Experience')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Converted Compensation Yearly')

# Create scatter plot for 'ConvertedCompYearly' vs. 'JobSatPoints_1'
plt.subplot(1, 2, 2)
sns.scatterplot(x='JobSatPoints_1', y='ConvertedCompYearly', data=df, color='red')
plt.title('Converted Compensation vs. Job Satisfaction')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Converted Compensation Yearly')

# Show the plots
plt.tight_layout()
plt.show()
```
