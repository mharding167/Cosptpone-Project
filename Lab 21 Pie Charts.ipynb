### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here


# Clean the data: Remove any rows where the 'DatabaseWantToWorkWith' column is missing
df.dropna(subset=['DatabaseWantToWorkWith'], inplace=True)

# Count occurrences of each database
database_counts = df['DatabaseWantToWorkWith'].value_counts()

# Get the top 5 most desired databases
top_5_databases = database_counts.head(5)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases.values, labels=top_5_databases.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Add a title
plt.title("Top 5 Most Desired Databases")

# Display the chart
plt.show()

```

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Sample data (assuming df is already loaded)
# df = pd.read_csv('your_data.csv')  # Uncomment to load your actual data

# Clean the data: Remove rows where the 'LanguageAdmired' column is missing
df.dropna(subset=['LanguageAdmired'], inplace=True)

# Split the 'LanguageAdmired' column into individual languages and count their occurrences
admired_languages = df['LanguageAdmired'].str.split(',').explode().str.strip()

# Count the occurrences of each admired language
admired_language_counts = admired_languages.value_counts()

# Get the top 5 most admired languages
top_5_admired_languages = admired_language_counts.head(5)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_admired_languages.values, labels=top_5_admired_languages.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Add a title
plt.title("Top 5 Most Admired Programming Languages")

# Display the chart
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here

# Search for a column containing 'AITool' in the name
print([col for col in df.columns if 'AITool' in col])

# Clean the data: Remove rows where the 'AIToolCurrently Using' column is missing
df.dropna(subset=['AIToolCurrently Using'], inplace=True)

# Split the 'AIToolCurrently Using' column into individual tools and count their occurrences
tools = df['AIToolCurrently Using'].str.split(',').explode().str.strip()

# Count the occurrences of each tool
tool_counts = tools.value_counts()

# Get the top 5 most used AI tools
top_5_tools = tool_counts.head(5)

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_5_tools.values, labels=top_5_tools.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Add a title
plt.title("Top 5 AI Tools Currently Used by Developers")

# Display the chart
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here

# Clean the data: Remove rows where the 'WebframeWantToWorkWith' column is missing
df.dropna(subset=['WebframeWantToWorkWith'], inplace=True)

# Split the 'WebframeWantToWorkWith' column into individual frameworks and count their occurrences
frameworks = df['WebframeWantToWorkWith'].str.split(',').explode().str.strip()

# Count the occurrences of each framework
framework_counts = frameworks.value_counts()

# Get the top 5 most desired web frameworks
top_5_frameworks = framework_counts.head(5)

# Create a list of colors for the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']  # List of 5 colors

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_frameworks.values, labels=top_5_frameworks.index, autopct='%1.1f%%', startangle=90, colors=colors)

# Add a title
plt.title("Top 5 Web Frameworks Developers Want to Work With")

# Display the chart
plt.show()
```
